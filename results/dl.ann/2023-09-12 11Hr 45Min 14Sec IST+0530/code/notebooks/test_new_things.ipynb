```python
num_classes = 7
true_class = np.random.randint(num_classes)
true_class_prob = np.random.rand()

random_numbers = np.random.rand(num_classes - 1)
remaining_prob = 1 - true_class_prob

false_class_prob = random_numbers / random_numbers.sum() * remaining_prob

prediction_prob = np.insert(false_class_prob, true_class, true_class_prob)
prediction_prob.sum()
```
